```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
```

```python
df = pd.read_excel('/content/drive/MyDrive/SkillBox/Как врать при помощи статистики. Практическая работа 1.xlsx')
```

```python
df = df.fillna(0)
```

Разброс между выручкой отчетной компании и другими достигался тем, что выручка первого года Shmuber составляла всего лишь 5. Не совсем честно сравнивать остальные года с этим показателем.

В xlsx файле изменила формулу расчета rate - сравниваю показатель выручки с **предыдущим** годом.

Теперь разброс устранен, к 2015 году компании близки в показателях.

```python
fix, ax = plt.subplots(figsize=(18,6))
sns.lineplot(data=df, x='year', y='rate', hue='company', ax=ax)
plt.ylabel("%")
```
