```python
import pandas as pd
from datetime import datetime, time, date
from dateutil.relativedelta import relativedelta
```

Готовим датафреймы:



```python
course_contents = pd.read_csv('/content/drive/MyDrive/SkillBox/Data/course_contents.csv')
courses = pd.read_csv('/content/drive/MyDrive/SkillBox/Data/courses.csv', index_col=0)
progress_phases = pd.read_csv('/content/drive/MyDrive/SkillBox/Data/progress_phases.csv',  parse_dates=['start_date', 'finish_date'], dayfirst=False)
progresses = pd.read_csv('/content/drive/MyDrive/SkillBox/Data/progresses.csv')
students = pd.read_csv('/content/drive/MyDrive/SkillBox/Data/students.csv', parse_dates=['birthday'])
```

```python
merged_courses = courses.merge(course_contents, left_on='id', right_on='course_id', how='left').drop(columns=['id'])
merged_progresses = progresses.merge(progress_phases, left_on='id', right_on='progress_id', how='left').drop(columns=['id'])
merged_df = pd.merge(merged_courses, merged_progresses, on = ['module_number', 'lesson_number', 'course_id'], how ='left')
merged_df.info()
```

```python
merged_w_students = pd.merge(merged_df, students, left_on = 'student_id', right_on='id', how ='left').drop(columns=['id'])
merged_w_students.info()
```

количество уроков в каждом модуле на каждом курсе;
медианное количество уроков в модуле на каждом курсе;
количество учеников на каждом курсе;
минимальный, максимальный, средний, медианный возраст студентов *(чтобы отсечь некорректные значения, в данном случае достаточно установить правдоподобный диапазон на ваше усмотрение)*;
минимальный, максимальный, средний, медианный возраст студентов на каждом курсе.

```python
# общее количество курсов в датасете
len(merged_df['course_id'].unique())
```

```python
# количество модулей на каждом курсе
merged_df.groupby('title')['module_number'].nunique()
```

```python
# количество уроков в каждом модуле на каждом курсе
merged_df.groupby(['title', 'module_number'])['lesson_number'].nunique().reset_index()
```

```python
# медианное количество уроков в модуле на каждом курсе
merged_df.groupby(['title', 'module_number'])['lesson_number'].nunique().reset_index().groupby('title')['lesson_number'].median()
```

```python
# количество учеников на каждом курсе
merged_w_students.groupby('title')['student_id'].nunique()
# тут не уверена, верно посчитала?
```

```python
# минимальный, максимальный, средний, медианный возраст студентов

merged_w_students['birthday'] = pd.to_datetime(merged_w_students['birthday'], yearfirst=True, format='%Y-%m-%d', errors='coerce')

merged_w_students['age'] = (datetime.today() - merged_w_students['birthday']).astype('<m8[Y]')
merged_w_students = merged_w_students[(merged_w_students['age'] >= 14) &  (merged_w_students['age'] <= 80)]

merged_w_students['age'].agg(['min', 'max', 'mean', 'median'])
```

```python
# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
merged_w_students.groupby('title').agg({'age': ['min', 'max', 'mean', 'median']})
```
